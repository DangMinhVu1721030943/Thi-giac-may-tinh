```python
img = load_image(my_path)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (100,75,150,100)
# Cắt vùng ảnh
croped_image = img.crop(region)
display(croped_image)
```

```python
# Dán
# Xác định vị trí cần dán
paste_position = (150,150)              
img.paste(croped_image,paste_position)
display(img)
```
