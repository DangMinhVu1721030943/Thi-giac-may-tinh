```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = r'C:\Users\dangminhvu\Pictures/abc.PNG'
```

```python
im = Image.open(my_path)
plt.imshow(im)
```

```python
im_gray =im.convert('L')
plt.imshow(im_gray,cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(), bins = 128)
plt.title('Bieu do gia tri histogram cua anh')
plt.xlabel('gia tri cua pixel')
plt.ylabel('so luong')
plt.show()
```

```python
# Làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

```python
# Đường dẫn ảnh
my_path = r'C:\Users\dangminhvu\Pictures/anh3.PNG'

# Đọc ảnh và hiển thị ảnh với trục tọa độ
im = Image.open(my_path).convert("L")

# Hiển thị ảnh trên trục tọa độ
plt.imshow(im)

# Tạo một ình vẽ mới
plt.figure()

# Đặt màu sắc là xám
plt.gray()

# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(im, origin="image")

# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')

# hiển thị ảnh
plt.show()
```
