```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = r'C:\Users\dangminhvu\Pictures/abc.PNG'
```

```python
im = Image.open(my_path)
plt.imshow(im)
```

```python
im_gray =im.convert('L')
plt.imshow(im_gray,cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# To mt hnh v mi
plt.figure()
plt.hist(im_array.flatten(), bins = 128)
plt.title('Bieu do gia tri histogram cua anh')
plt.xlabel('gia tri cua pixel')
plt.ylabel('so luong')
plt.show()
```

```python
# Lm ni bt ng bin
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
